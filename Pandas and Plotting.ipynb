<h1>Pandas</h1>
<p>The cell below is the code that answers the questions of day #4</p>

```python
import numpy as np
import pandas as pd

DataFrame=pd.read_csv("EnergyCostSavings.csv")
   
print("There are %d different company names in the data set." % len(np.unique(DataFrame["Company Name"].values)))
print("There was %d jobs created in Queens." % np.sum(DataFrame[(DataFrame["City"]=="Queens") & (DataFrame["Job created"].notna())]["Job created"].values))

Domains=set()

for Email in DataFrame[DataFrame["company email"].notna()]["company email"].values:
    Domains.add(Email.partition('@')[2])
    
print("There are %d different domain names." % len(Domains))

Dic={}

for Item in DataFrame[DataFrame["NTA"].notna()]["NTA"].values:
    Dic[Item]=Dic.get(Item,0)+1

Data={}

for Item in Dic:
    if Dic[Item]>=5:
        Jobs=0
        
        for Value in DataFrame[DataFrame["NTA"]==Item]["Job created"].values:
            if not np.isnan(Value):
                Jobs+=Value
                
        Data[Item]=[np.average(DataFrame[DataFrame["NTA"]==Item]["Total Savings"].values),Jobs]
        
DF=pd.DataFrame(Data)
print(DF)
DF.to_csv("My-csv-file.csv",sep=' ')
```

<h1>Plotting</h1>
<p>The cell below is the code that answers the questions of day #5</p>

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

DataFrame=pd.read_csv("EnergyCostSavings.csv")
   
Dic={}

for Item in DataFrame[DataFrame["NTA"].notna()]["NTA"].values:
    Dic[Item]=Dic.get(Item,0)+1

Data1={}

for Item in Dic:
    if Dic[Item]>=5:
        Jobs=0
        
        for Value in DataFrame[DataFrame["NTA"]==Item]["Job created"].values:
            if not np.isnan(Value):
                Jobs+=Value
                
        Data1[Item]=[np.average(DataFrame[DataFrame["NTA"]==Item]["Total Savings"].values),Jobs]
        
x1=np.array([0])
y1=np.array([0])

for Item in Data1:
    x1=np.append(x1,Data1[Item][0])
    y1=np.append(y1,Data1[Item][1])
    
Index=0
x2=np.array([0])
y2=np.array([0])
    
for i in DataFrame.index:
    Index+=1
    
    if DataFrame["Job created"][i]==DataFrame["Job created"][i]:
        x2=np.append(x2,Index)
        y2=np.append(y2,DataFrame["Job created"][i])

Figure,Axis=plt.subplots(3,1)

Axis[0].set_title("Scatter plot")
Axis[0].plot(x1,y1,'o')
Axis[0].set_xlabel("Average savings")
Axis[0].set_xscale('log')
Axis[0].set_ylabel("Jobs created")

Axis[1].set_title("Histogram")
Axis[1].hist(x1)
Axis[1].set_xlabel("Average savings")
Axis[1].set_ylabel("Frequency (log)")
Axis[1].set_yscale('log')

Axis[2].set_title("Line plot")
Axis[2].plot(x2,y2)
Axis[2].set_xlabel("Month index")
Axis[2].set_ylabel("Jobs created")
plt.show()
```
